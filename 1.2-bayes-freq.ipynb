### Likelihood

似然是一种概率，只是它描述了在给定某个假设参数的情况下已经观察到的数据的概率。一般来说，概率是指未来事件发生的可能性。在这里区分概率和似然是必要的。概率针对的是给定固定假设参数的可变数据。似然是针对固定观察数据的可变假设参数。如果 'x' 代表我们的数据，\\(\theta\\) 代表我们的假设参数，则似然写为

$$ Likelihood(\theta | data) = P(x | \theta) $$

值得指出的是，似然不是有效的概率，因为通过观察数据改变 theta 而形成的概率密度函数积分不为 1。正是出于这个原因，术语似然存在，即提醒我们它不是一个有效的概率密度。

### Inference 

推理是指识别代表我们假设的参数的分布的过程。然而，这是对推理的明显贝叶斯定义，因为频率论者相信真实参数的存在。我们将在接下来花一些时间来理解这些差异。后验分布是在推理步骤中计算的，可以表示为

$$ P(\theta | x) $$

## The Bayesian and the Frequentist

在频率论统计中，我们的概率仅基于我们所观察到的。

而贝叶斯解释概率为 对一个事件相信程度的衡量，也就是说对一件事情发生我们有多大的信念，而这其实是对概率的一种很自然的解释。

频率学派从事件长期发生频率中来得到概率，但是很多时候我们并没有长期的记录，很多不确定事件只有很少几次记录。为了处理这种情况，贝叶斯学派将概率解释为 对一个事件发生的信念的衡量 measure of belief/confidence of an event occurring。简单地说，概率就是观点的总结，对一个事发生完全没信心，那么对信念的衡量就是0，充满信心，认为铁定发生，那么对信念的衡量就是1，零一之间就是对不同的结果发生都有一些信心。这和频率学派的概率也是一致的，比如观察到一个事件的频率，那么在没有其他外部信息的条件下，个人的信念就应该等于这个频率。那么根据概率等于信念的定义，对小样本事件，概率也可以很有意义，我们对一件事的信心有多大。

这里值得注意的是，信心被赋予了“我们”自己，而不是“自然”。所以对同一件事，由于背景信息的不同，不同的人会有不同的看法，比如一位内科病人表现出症状 X， y 和 z。有多种疾病可能导致所有这些症状，但仅存在一种疾病是真正导致病人生病的原因。一个医生会认为是一种疾病导致，但是第二位医生的信念可能就会略有不同。

贝叶斯就是这样将信念视为概率的哲学。当我们与世界互动时，我们会不断采用它，虽然只看到部分事实，但我们会收集证据以形成信念。为了与传统的概率表示法保持一致，我们把对一个事件A的信念表示为 P（A）。我们称这个量为先验概率。当看到更多证据后我们会更新信念。即使是证据与最初认为的相反，也不能忽略该证据。我们将更新后的信念表示为P（A|X），解释为给定证据 X后对事件A的信念。并称更新后的信念为后验概率，以便将其与先验概率进行对比。比如，P（A ）：病人可能患有多种疾病。 P（A|X）： 进行血液检查产生证据 X，从而考虑排除一些可能的疾病。

我们在看到新证据后没有完全放弃先前的信念 X，但是重新加权了先前的对不同事件信念的加权，以纳入新的证据（即，相比于与其他信念，我们对某些信念给予了更大的权重或信心）。通过引入事件的先验不确定性，我们已经承认我们所做的任何猜测都可能是非常错误的。观察数据，证据或其他信息后，我们更新信念，那么我们的猜测也就会更少有错。

简单总结，在贝叶斯统计中，我们使用我们对环境或产生观察结果的系统的先验理解来推断未来事件的可能性。正如本·兰伯特在他的《贝叶斯统计学生指南》一书中所说，“对于贝叶斯主义者来说，为了定义概率，事件没有必要是可重复的。它们只是帮助表达我们不确定性的抽象概念”。贝叶斯观点和频率论观点之间的根本区别取决于人们如何对待参数和数据。对于频率学派者来说，与分配给事件的概率相关的不确定性来自随机性（任意不确定性），他们没有考虑因缺乏知识而产生的不确定性（认知不确定性）。

### The Bayesian

数据是固定的，我们根据新数据更新我们的信念。某个参数所代表的信念是我们更新的随机变量。贝叶斯者觉得 我们对这些参数的了解随着时间的推移而发展的想法 是合理的。在存在可以估计的真实参数值的情况下，贝叶斯定理允许人们指定该参数的不确定性。

### The Frequentist

对于频率主义者，数据是从随机过程中抽样的结果。他们认为数据是变化的，而产生数据的这个随机过程的参数是固定的。他们认为这个参数是无数次实验的平均值。这种方法对于不能重复的事件是个问题。

### The Differences Explained

本·兰伯特 (Ben Lambert) 的美国大选例子有助于说明这种微妙的差异。

**Bayesian** - 民主党胜选的概率是 0.75.

**Frequentist** - 由于在特定选举中只能获得一个样本，因此这不是一项可重复的练习，我们无法从此次选举的总体结果中抽样。

再例如，在重复 5 次出现 4 个正面的硬币翻转中，对于频率者，出现正面的概率为 4/5，代表从抛硬币得到正面或反面的生成过程的分布可以由参数 \\(\theta\\) 表示。当我们想要确定“n”次抛硬币的正面次数时，可以使用二项分布对其进行建模。如果“O”代表观察到的结果，“H”代表正面的结果，那么似然可以写为

$$Likelihood = P(O=H/ \; \theta) = 0.8$$ 

频率论者将此用作参数\\(\theta\\) 的概率。

然而，在贝叶斯方法中，当出现正面时，估计参数 $\theta$ 的后验概率，会假设先验概率 $P(\theta)$ = 0.5

$$ Posterior \propto Likelihood \times Prior \longrightarrow P(\theta | O = H) \propto P( O = H | \theta) \times P(\theta) $$

$$ Posterior \propto 0.8 * 0.5 $$

**总而言之：在推理中，我们想要衡量给定某些数据的时候假设的概率。频率论者选择一个假设并决定给定这个假设的时候数据的概率。然后他们将其用作假设的证据。贝叶斯学派更进一步将其反转以获得给定数据的假设的概率。**

假设正在测量您所在地区的公民的身高，并且有一个**假设**他们的身高正态分布，平均值为 5.75 英尺，标准差为 0.4 英尺。我们观察到一个身高 6.5 英尺的人。频率论者必须计算这种观察发生的概率为

$$P(h = 6.5 | N(\mu, \sigma) )$$ 

并使用阈值来确定我们的原始假设是否有效。在贝叶斯统计中，鉴于我们观察到高度为 6.5 英尺的样本，后验是我们假设参数的概率分布

$$P(N(\mu, \sigma) | h = 6.5$$

然而，在这种情况下，我们使用主观先验来计算这个后验。我们用频率统计中的任意阈值（通常为 1% 或 5%，取决于正在解决的问题）与贝叶斯统计中的主观先验进行交换。有人可能会争辩说，这两种方法都是将领域知识和主观专业知识纳入决策过程的不同方式。然而，贝叶斯不需要接受或拒绝假设，因为它们具有作为后验结果的完整分布，因此可以用假设量化不确定性。有关假设检验的更多信息，请参见后续部分。

### Features of Bayesian Inference 

典型的例子是参数的贝叶斯推断。假设一个模型，其中数据 x 是根据未知参数 θ 从概率分布生成的。我们还假设我们有关于参数 θ 的先验知识，可以表示为概率分布 p(θ)。然后，当观察到数据 x 时，我们可以使用贝叶斯定理更新关于该参数的先验知识。如下所示：

![](pictures/1_04pd7c6QIHXYHgAelzzWlg@2x.png)

所以贝叶斯推断一般步骤是：

1. 为假设（后验）和数据（似然）分配概率

2. 通过制定“主观”先验来利用专家知识。先验的使用一直是争论的焦点，但是当明确说明这一点时，它允许每个人理解和挑战结果背后的假设，也允许改进先验。

3. 计算后验还需要p(x)，$p(x)=\int_\theta p(x|\theta)p(\theta)d\theta$，多个参数的时候就需要对多个参数进行积分，计算后验的计算成本可能很高。很多时候，我们不得不求助于近似技术，因为与贝叶斯统计中的后验计算相关的积分无法解析计算。使用了许多近似技术。

    1. 拉普拉斯近似

    2. 变分逼近

    3. 蒙特卡罗技术（本repo关注的，后面会介绍的主题）

    4. 消息传递算法

为什么使用贝叶斯推断？

#### The Bayesian landscape

当我们设置N个未知数的贝叶斯推断问题时，我们隐式地创建了一个 N维的存在先验分布的空间。与空间相关联的是一个附加维度，我们可以将其描述为位于空间顶部的曲面或曲线，它反映了特定点的先验概率。空间的表面由我们的先验分布定义。例如，如果我们有两个未知数p1 和 p2，并且两者的先验是 Uniform（0 ，5 ），那么创建的空间是一个长度为5的正方形，并且表面是一个位于正方形顶部的平面（表示每个点的可能性均等）。

```python
%matplotlib inline
import scipy.stats as stats
from IPython.core.pylabtools import figsize
import numpy as np
figsize(12.5, 4)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

jet = plt.cm.jet
fig = plt.figure()
x = y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)

plt.subplot(121)
uni_x = stats.uniform.pdf(x, loc=0, scale=5)
uni_y = stats.uniform.pdf(y, loc=0, scale=5)
M = np.dot(uni_x[:, None], uni_y[None, :])
im = plt.imshow(M, interpolation='none', origin='lower',
                cmap=jet, vmax=1, vmin=-.15, extent=(0, 5, 0, 5))

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Landscape formed by Uniform priors.")

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(X, Y, M, cmap=plt.cm.jet, vmax=1, vmin=-.15)
ax.view_init(azim=390)
plt.title("Uniform prior landscape; alternate view");
```

或者，如果两个先验是 Exp（3 ） 和 Exp（10 ），则该空间在二维平面上均为正数，先验的surface看起来像是从点（0,0）开始流过正数的瀑布。

下图将其可视化。颜色越深越红，分配给该位置的先验概率就越高。相反，蓝色较深的区域表示我们的先验对该位置分配的概率非常低。

```python
figsize(12.5, 5)
fig = plt.figure()
plt.subplot(121)

exp_x = stats.expon.pdf(x, scale=3)
exp_y = stats.expon.pdf(x, scale=10)
M = np.dot(exp_x[:, None], exp_y[None, :])
CS = plt.contour(X, Y, M)
im = plt.imshow(M, interpolation='none', origin='lower',
                cmap=jet, extent=(0, 5, 0, 5))
#plt.xlabel("prior on $p_1$")
#plt.ylabel("prior on $p_2$")
plt.title("$Exp(3), Exp(10)$ prior landscape")

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(X, Y, M, cmap=jet)
ax.view_init(azim=390)
plt.title("$Exp(3), Exp(10)$ prior landscape; \nalternate view");
```

这些是2D空间中的简单示例，我们的大脑可以很好地理解曲面。实际上，我们的先验产生的空间和曲面的维度可能要大得多。

如果这些表面描述了我们在未知数上的先验分布，那么在融合观测数据X之后，空间将会发生什么？数据X不会改变空间，但会通过伸缩先验曲面的构造来改变空间曲面以反映真实参数可能存在的位置。更多的数据意味着更多的伸缩，并且与新形成的形状相比，我们的原始形状会变得微不足道。数据越少，我们保留原始形状就更多。无论如何，所得曲面都描述了后验分布。

不幸的是，我们不可能在高维可视化它。对于二维而言，数据实质上是推高原始表面以形成高耸的山脉。可以通过先验概率分布观测数据上升的趋势检查某些区域的后验概率。先验概率越小意味着阻力越大。因此，在上述双指数先验情况下，在（0,0）角附近喷出一座山（或多座山）的可能将比在接近（5,5）喷出山的可能高得多，因为（5,5）附近的低先验概率表明阻力更大。峰值反映了可能找到真实参数的后验概率。值得注意的是，如果先验概率已分配为0，则不会在此分配后验概率。

假设上述先验代表两个泊松分布不同的参数 λ。现在我们观察到一些数据点并重新可视化整个landscape（TODO: 这块似乎有点问题，但是主要是为了明确逻辑，里面具体代码细节不用管）：

```python
# create the observed data

# sample size of data we observe, trying varying this (keep it less than 100 ;)
N = 1

# the true parameters, but of course we do not see these values...
lambda_1_true = 1
lambda_2_true = 3

#...we see the data generated, dependent on the above two values.
data = np.concatenate([
    stats.poisson.rvs(lambda_1_true, size=(N, 1)),
    stats.poisson.rvs(lambda_2_true, size=(N, 1))
], axis=1)
print("observed (2-dimensional,sample size = %d):" % N, data)

# plotting details.
x = y = np.linspace(.01, 5, 100)
likelihood_x = np.array([stats.poisson.pmf(data[:, 0], _x)
                        for _x in x]).prod(axis=1)
likelihood_y = np.array([stats.poisson.pmf(data[:, 1], _y)
                        for _y in y]).prod(axis=1)
L = np.dot(likelihood_x[:, None], likelihood_y[None, :])
```

```python
figsize(12.5, 12)
# matplotlib heavy lifting below, beware!
plt.subplot(221)
uni_x = stats.uniform.pdf(x, loc=0, scale=5)
uni_y = stats.uniform.pdf(x, loc=0, scale=5)
M = np.dot(uni_x[:, None], uni_y[None, :])
im = plt.imshow(M, interpolation='none', origin='lower',
                cmap=jet, vmax=1, vmin=-.15, extent=(0, 5, 0, 5))
plt.scatter(lambda_2_true, lambda_1_true, c="k", s=50, edgecolor="none")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Landscape formed by Uniform priors on $p_1, p_2$.")

plt.subplot(223)

plt.contour(x, y, M * L)
im = plt.imshow(M * L, interpolation='none', origin='lower',
                cmap=jet, extent=(0, 5, 0, 5))
plt.title("Landscape warped by %d data observation;\n Uniform priors on $p_1, p_2$." % N)
plt.scatter(lambda_2_true, lambda_1_true, c="k", s=50, edgecolor="none")
plt.xlim(0, 5)
plt.ylim(0, 5)

plt.subplot(222)
exp_x = stats.expon.pdf(x, loc=0, scale=3)
exp_y = stats.expon.pdf(x, loc=0, scale=10)
M = np.dot(exp_x[:, None], exp_y[None, :])

plt.contour(x, y, M)
im = plt.imshow(M, interpolation='none', origin='lower',
                cmap=jet, extent=(0, 5, 0, 5))
plt.scatter(lambda_2_true, lambda_1_true, c="k", s=50, edgecolor="none")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Landscape formed by Exponential priors on $p_1, p_2$.")

plt.subplot(224)
# This is the likelihood times prior, that results in the posterior.
plt.contour(x, y, M * L)
im = plt.imshow(M * L, interpolation='none', origin='lower',
                cmap=jet, extent=(0, 5, 0, 5))

plt.scatter(lambda_2_true, lambda_1_true, c="k", s=50, edgecolor="none")
plt.title("Landscape warped by %d data observation;\n Exponential priors on \
$p_1, p_2$." % N)
plt.xlim(0, 5)
plt.ylim(0, 5);
```

左边的图是 均匀（0 ，5 ）先验的 变形，右图是具有指数先验的变形。请注意，尽管两种情况下观察到的数据相同，但后验看起来彼此不同。

请注意，指数先验（右下图）在该图右上角的值上只放置了很少的后验权重：这是因为先验值在此处没有太大的权重。另一方面，均匀先验很乐意将后后验权重放在右上角，因为先验在右上角放置更多的权重。

还要注意，在指数情况下，最高点（对应最暗的红色）朝（0,0）倾斜，这是指数先验将更多先验权重放在（0,0）角的结果。

黑点代表真实参数。即使只有1个采样点，山脉也会尝试包含true参数。当然，采用1个样本大小进行推断是非常简单的，选择如此小的样本大小仅是说明性的。

### Features of Frequentist Inference

1. 假设本身是没有概率的（无后验）

2. 基于NHST（null hypothesis significance testing）

3. 没有使用先验，但通过使用 p 值仍然存在主观性

4. 计算密集度较低

5. 频率学方法中使用的最大似然方法的一个缺点是模型容易过度拟合。

#### Hypothesis Testing

使用显着性检验的假设检验是通过制定零假设并在此零假设下测试事件发生的概率来执行的。如果这个概率小于预定义的阈值（通常为 5%，但非常依赖于上下文），我们拒绝原假设，否则我们说事件在原假设下有合理的机会发生。零假设背后的一般概念是，这是一个我们想要证明错误的假设，与零假设相反的是备择假设。 P 值和置信区间通常用于研究中以传达实验的可信度。但是，这些 p 值取决于 [实验设置](https://en.wikipedia.org/wiki/Likelihood_principle#Experimental_design_arguments_on_the_likelihood_principle).

假设我们对弗吉尼亚州 2020 年 12 月的温度进行了一组测量。结果表明平均温度为 45F。从 1980 年到 2020 年的所有年份的平均温度为 50F。我们想使用具有 p 值的 NHST 来测试这是否是异常。我们的零假设是这不是异常现象，替代假设是这是一个异常寒冷的 12 月。我们必须选择一个阈值来接受或拒绝我们的假设，如上所述，这个阈值通常选择为 5%。为了证实我们的怀疑，我们必须拒绝原假设。我们首先获得数据在零假设下发生的概率，即给定 1980 年至 2020 年平均气温分布的 2020 年平均气温的概率。如果这个概率小于 5%，我们拒绝零假设说2020 年的数据不太可能来自原假设分布。因此，这是一个比弗吉尼亚习惯看到的更冷的冬天。如果概率大于 5%，我们不能拒绝原假设，因为有一些理由相信（基于我们的主观阈值）这些数据可能是由原假设分布产生的。因此，它不会比预期的更冷，因此不是异常。
